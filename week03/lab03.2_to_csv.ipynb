```python
# Read hospitals.json to inspect its structure
df_json = (
    spark.read
        .option("multiline", "true")
        .json("/Workspace/Users/pboal@wustl.edu/data-5035-2026/week03/hospitals.json")
)
```

```python
df_json.printSchema()
```

```python
from pyspark.sql.functions import explode

hospitals_df = (
    df_json
    .select("system", explode("hospitals").alias("hospital"))
    .select(        
        "system",
        "hospital.name",
        "hospital.city",
        "hospital.beds")
)
```

```python
hospitals_df
```

```python
(hospitals_df
 .write
 .mode("overwrite")
 .option("header", "true")
 .csv("/Volumes/workspace/default/data_5035/week03/hospitals_csv"))
```
