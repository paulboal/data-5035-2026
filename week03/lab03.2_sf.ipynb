```python
import json
```

```python
regional_hospitals = """
[
    {
        "system": "BJC",
        "hospitals": [
            { "name": "BJH", "city": "St. Louis", "beds": 1432 },
            { "name": "MOBap", "city": "Creve Coeur", "beds": 1107 }
        ]
    },
    {
        "system": "Mercy",
        "hospitals": [
            { "name": "Mercy STL", "city": "Creve Coeur", "beds": 983 }
        ]
    },   
    {
        "system": "SSM",
        "hospitals": [
            { "name": "SLUH", "city": "St. Louis", "beds": 965 }
        ]
    }    
]
"""
```

```python
h = json.loads(regional_hospitals)
```

```python
h
```

```python
# Reorganize hospital data: group by city instead of by system
cities = {}  # Dictionary to hold hospitals grouped by city

# Loop through each hospital system
for system in h:
    system_name = system['system']  # Get the system name (e.g., 'BJC', 'Mercy')
    
    # Loop through each hospital in this system
    for hospital in system['hospitals']:
        city = hospital['city']
        
        # Initialize empty list for city if not seen before
        if city not in cities:
            cities[city] = []
        
        # Add hospital info to the city's list
        cities[city].append({
            'name': hospital['name'],
            'system': system_name,
            'beds': hospital['beds']
        })

cities
```
