```python
import json
```

```python
regional_hospitals = """
[
    {
        "system": "BJC",
        "hospitals": [
            { "name": "BJH", "city": "St. Louis", "beds": 1432 },
            { "name": "MOBap", "city": "Creve Coeur", "beds": 1107 }
        ]
    },
    {
        "system": "Mercy",
        "hospitals": [
            { "name": "Mercy STL", "city": "Creve Coeur", "beds": 983 }
        ]
    },   
    {
        "system": "SSM",
        "hospitals": [
            { "name": "SLUH", "city": "St. Louis", "beds": 965 }
        ]
    }    
]
"""
```

```python
h = json.loads(regional_hospitals)
```

```python
h
```

```python
# Initialize an empty dictionary to group hospitals by city
city_hospitals = {}

# Iterate over each health system in the list
for system_info in h:
    system = system_info["system"]  # Get the system name
    # Iterate over each hospital in the current system
    for hospital in system_info["hospitals"]:
        city = hospital["city"]  # Get the city of the hospital
        # Create a dictionary entry for the hospital with its name, system, and bed count
        hospital_entry = {
            "name": hospital["name"],
            "system": system,
            "beds": hospital["beds"]
        }
        # Add the hospital entry to the list for its city, creating the list if needed
        city_hospitals.setdefault(city, []).append(hospital_entry)

# Display the dictionary of hospitals grouped by city
display(city_hospitals)
```
